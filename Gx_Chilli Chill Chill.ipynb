# **1. State Problem**


### **1.1 Problem and Pain point**
สำหรับผู้ค้ารายย่อยหรือแม่ค้าในตลาดสด การคัดแยกพริกที่รับมาจำหน่ายต่อเป็นขั้นตอนสำคัญที่ส่งผลต่อรายได้และความน่าเชื่อถือของร้านโดยตรง แต่กระบวนการนี้มักประสบปัญหาที่มองข้ามไป:

*   **การคัดแยกที่เสียเวลา** แม่ค้าต้องใช้เวลา considerable ในการคัดแยกพริกทีละเม็ดด้วยสายตา เพื่อแยกพริกที่ยังไม่สุก, พริกที่เริ่มเน่าเสีย หรือพริกที่คุณภาพไม่ดีออกจากกัน ซึ่งเวลาที่เสียไปนี้สามารถนำไปใช้ในการบริการลูกค้าหรือจัดการร้านได้
*   **มาตรฐานไม่นิ่งและเกิดของเสีย** การคัดแยกด้วยสายตาอาจเกิดความผิดพลาด ทำให้มีพริกที่ไม่ได้คุณภาพปะปนไปขายให้ลูกค้า ซึ่งส่งผลต่อชื่อเสียงของร้านในระยะยาว นอกจากนี้ การไม่สามารถคัดแยกพริกที่ใกล้เน่าเสียออกได้ทันท่วงที ทำให้เกิดการสูญเสีย (Waste) ของสินค้าที่ต้องทิ้งไปโดยเปล่าประโยชน์
*   **การตั้งราคาที่ไม่เหมาะสม** พริกที่มีระดับความสุกต่างกัน (เช่น พริกเขียว, พริกแดง) ควรมีราคาต่างกัน การคัดแยกที่รวดเร็วและแม่นยำจะช่วยให้แม่ค้าสามารถตั้งราคาขายตามคุณภาพของสินค้าได้อย่างเหมาะสม เพิ่มโอกาสในการทำกำไร

ดังนั้น การมีเครื่องมือหรือแอปพลิเคชันง่ายๆ ที่ช่วยจำแนกระดับความสุกของพริกได้อัตโนมัติ จะเป็นเหมือนผู้ช่วยคนสำคัญที่ช่วยให้แม่ค้าประหยัดเวลา, ลดปริมาณของเสีย, และสร้างมาตรฐานคุณภาพให้กับสินค้าของตนเองได้
### **1.2 Related work**
การนำเทคโนโลยีคอมพิวเตอร์วิทัศน์ (Computer Vision) มาประยุกต์ใช้ในภาคการเกษตรมีมาอย่างต่อเนื่อง โดยเฉพาะอย่างยิ่งในงานด้านการตรวจจับและคัดแยกผลผลิตทางการเกษตร ซึ่งสามารถแบ่งแนวทางการพัฒนาได้ดังนี้


* **ยุคเริ่มต้น (Traditional Image Processing)**
ในช่วงแรก งานวิจัยมักใช้เทคนิคการประมวลผลภาพแบบดั้งเดิม เช่น การวิเคราะห์ค่าสีในปริภูมิสีต่างๆ (RGB, HSV) เพื่อแยกแยะความแตกต่างของสีผลไม้ แต่เทคนิคเหล่านี้มีข้อจำกัดอย่างมากเมื่อต้องเจอกับสภาพแสงที่แตกต่างกัน, เงา, หรือการมีวัตถุอื่นบดบัง
*   **ยุค Machine Learning** ต่อมามีการใช้โมเดล Machine Learning แบบดั้งเดิม เช่น Support Vector Machines (SVM) ร่วมกับการสกัดคุณลักษณะเด่น (Feature Extraction) ด้วยมือ เช่น รูปทรง, พื้นผิว ซึ่งให้ผลลัพธ์ที่ดีขึ้น แต่ยังคงมีความซับซ้อนและต้องอาศัยผู้เชี่ยวชาญในการออกแบบคุณลักษณะ
* **ยุค Deep Learning (ปัจจุบัน)**
ปัจจุบันเทคโนโลยี Deep Learning ได้เข้ามามีบทบาทสำคัญ โดยเฉพาะโมเดลในกลุ่ม Object Detection ซึ่งสามารถเรียนรู้และสกัดคุณลักษณะเด่นจากภาพได้โดยอัตโนมัติ ทำให้มีความแม่นยำและทนทานต่อสภาพแวดล้อมที่หลากหลายมากกว่า โดยโมเดลที่เป็นที่นิยมในงานลักษณะนี้แบ่งได้เป็น 2 กลุ่มหลัก:
     *  **One-Stage Detectors**
      เช่น โมเดลตระกูล YOLO (You Only Look Once) และ SSD ซึ่งมีจุดเด่นด้านความเร็วสูง เหมาะสำหรับงานที่ต้องการการประมวลผลแบบเรียลไทม์ แต่ความแม่นยำอาจไม่สูงเท่าโมเดลแบบ Two-Stage
     *   **Two-Stage Detectors** เช่น โมเดลตระกูล R-CNN, Fast R-CNN, และ Faster R-CNN โมเดลกลุ่มนี้มีกระบวนการทำงาน 2 ขั้นตอน คือการค้นหาบริเวณที่คาดว่าจะเป็นวัตถุ (Region Proposal) ก่อน แล้วจึงทำการจำแนกประเภท ทำให้มีความแม่นยำในการระบุตำแหน่งและประเภทของวัตถุสูงมาก
     
สำหรับโครงงานนี้ การเลือกใช้ Faster R-CNN ถือว่ามีความเหมาะสมอย่างยิ่ง เนื่องจากงานคัดแยกผลผลิตทางการเกษตรให้ความสำคัญกับ ความแม่นยำสูงสุด (High Accuracy) เพื่อให้ได้ผลผลิตที่มีคุณภาพตามมาตรฐาน ซึ่ง Faster R-CNN เป็นโมเดลที่ได้รับการยอมรับอย่างกว้างขวางว่ามีประสิทธิภาพด้านความแม่นยำเป็นเลิศ และเหมาะสมกับการนำมาใช้แก้ปัญหาในอุตสาหกรรมการเกษตรได้เป็นอย่างดี

## 2.4  การแปลงข้อมูล (Data Transformation)

```python
class CocoDataset(CocoDetection):
    def __init__(self, img_folder, ann_file, transform=None):
        super().__init__(img_folder, ann_file)
        self.transform = transform

    def __getitem__(self, idx):
        img, targets = super().__getitem__(idx)

        boxes, labels = [], []
        for obj in targets:
            xmin, ymin, w, h = obj["bbox"]
            xmax, ymax = xmin + w, ymin + h

            # กันกรณีที่ w หรือ h <= 0
            if w <= 0 or h <= 0:
                continue

            boxes.append([xmin, ymin, xmax, ymax])
            labels.append(obj["category_id"])

        # ถ้าไม่มี object → สร้าง tensor ว่าง
        if len(boxes) == 0:
            boxes = torch.zeros((0, 4), dtype=torch.float32)
            labels = torch.zeros((0,), dtype=torch.int64)
        else:
            boxes = torch.as_tensor(boxes, dtype=torch.float32)
            labels = torch.as_tensor(labels, dtype=torch.int64)

        target = {"boxes": boxes, "labels": labels}

        if self.transform:
            img = self.transform(img)

        return img, target
```

```python
transform = T.ToTensor()
train_data = CocoDataset(TRAIN_IMG, TRAIN_JSON, transform)
val_data   = CocoDataset(VAL_IMG, VAL_JSON, transform)

train_loader = DataLoader(train_data, batch_size=4, shuffle=True,
                          collate_fn=lambda x: tuple(zip(*x)))
val_loader   = DataLoader(val_data, batch_size=2, shuffle=False,
                          collate_fn=lambda x: tuple(zip(*x)))

print("Train set:", len(train_data), "| Val set:", len(val_data))
```

# 3. Train (การฝึกสอนโมเดล)

### 3.1 AI Model (Faster R-CNN)

```python
NUM_CLASSES = 4   # background=0, High=1, Medium=2, Low=3

model = fasterrcnn_resnet50_fpn(weights="DEFAULT")

in_feat = model.roi_heads.box_predictor.cls_score.in_features
model.roi_heads.box_predictor.cls_score = nn.Linear(in_feat, NUM_CLASSES)
model.roi_heads.box_predictor.bbox_pred = nn.Linear(in_feat, NUM_CLASSES * 4)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

### 3.2 Hyperparameters

```python
num_epochs = 20
params = [p for p in model.parameters() if p.requires_grad]

optimizer = torch.optim.SGD(
    model.parameters(),
    lr=0.005,            # learning rate ต้องใส่
    momentum=0.9,
    weight_decay=0.0005
)
```

### 3.3 Model Training


```python
from torchmetrics.detection import MeanAveragePrecision
metric = MeanAveragePrecision().to(device)

for epoch in range(num_epochs):
    # --- Training Step (เหมือนเดิม) ---
    model.train()
    total_train_loss = 0
    for imgs, tgts in train_loader:
        imgs_gpu = [img.to(device) for img in imgs]
        tgts_gpu = [{k: v.to(device) for k, v in t.items()} for t in tgts]
        loss_dict = model(imgs_gpu, tgts_gpu)
        losses = sum(loss for loss in loss_dict.values())
        optimizer.zero_grad()
        losses.backward()
        optimizer.step()
        total_train_loss += losses.item()
    avg_train_loss = total_train_loss / len(train_loader)

    # --- Validation Step (⭐️ อัปเดตใหม่ ⭐️) ---
    model.eval()
    total_val_loss = 0 # 1. สร้างตัวแปรสำหรับเก็บ val_loss
    with torch.no_grad():
        for imgs, tgts in val_loader:
            imgs_gpu = [img.to(device) for img in imgs]
            tgts_gpu = [{k: v.to(device) for k, v in t.items()} for t in tgts]

            # --- คำนวณ Validation Loss ---
            # 2. สลับไป .train() ชั่วคราวเพื่อเอา loss dict
            model.train()
            loss_dict_val = model(imgs_gpu, tgts_gpu)
            loss_val = sum(l for l in loss_dict_val.values())
            total_val_loss += loss_val.item()

            # 3. สลับกลับไป .eval() เพื่อเอา predictions
            model.eval()
            predictions = model(imgs_gpu)

            # --- อัปเดต mAP Metric ---
            metric.update(predictions, tgts_gpu)

    # 4. คำนวณค่าเฉลี่ยของ Val Loss
    avg_val_loss = total_val_loss / len(val_loader)

    # --- คำนวณ mAP และรีเซ็ต ---
    mAP_scores = metric.compute()
    metric.reset()

    # 5. อัปเดต print statement ให้แสดงผล val_loss ด้วย
    print(f"Epoch {epoch+1}/{num_epochs}, Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}, "
          f"mAP: {mAP_scores['map']:.4f}, mAP@.50: {mAP_scores['map_50']:.4f}")
```

```python
import torch
import torchvision
import torchvision.transforms as T
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from collections import Counter

# -------------------------
# ตั้งค่า
# -------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
num_classes = 4  # 👈 จำนวน class ของคุณ (รวม background)

# -------------------------
# โหลดโมเดล
# -------------------------
# ใช้สถาปัตยกรรมเดียวกับตอนเทรน
model = torchvision.models.detection.fasterrcnn_resnet50_fpn(weights=None, num_classes=num_classes)
model.load_state_dict(torch.load(r"C:\Users\User\Desktop\jupyter\thaichilli.pth", map_location=device))
model.to(device)
model.eval()

# -------------------------
# โหลดภาพและเตรียมข้อมูล
# -------------------------
transform = T.Compose([T.ToTensor()])
img_path = r"C:\Users\User\Desktop\jupyter\Chilli_Detect-8\valid\20251002_200708_jpg.rf.c3861bf75705f58ba2f07dd9a4b5a26f.jpg"
img = Image.open(img_path).convert("RGB")
img_tensor = transform(img).unsqueeze(0).to(device)

# -------------------------
# Run prediction
# -------------------------
with torch.no_grad():
    prediction = model(img_tensor)

boxes  = prediction[0]['boxes']
labels = prediction[0]['labels']
scores = prediction[0]['scores']

# -------------------------
# Plot Bounding Box
# -------------------------
threshold = 0.5
fig, ax = plt.subplots(1, figsize=(12, 8))
ax.imshow(img)

for box, label, score in zip(boxes, labels, scores):
    if score > threshold:
        x1, y1, x2, y2 = box.cpu().numpy()
        lbl = int(label.cpu().item())
        scr = float(score.cpu().item())

        rect = patches.Rectangle(
            (x1, y1), x2 - x1, y2 - y1,
            linewidth=2, edgecolor='red', facecolor='none'
        )
        ax.add_patch(rect)
        ax.text(x1, y1, f"Class {lbl} ({scr:.2f})",
                fontsize=10, color='white', backgroundcolor="red")

plt.title("Predicted Bounding Boxes")
plt.show()

# -------------------------
# นับจำนวน Class ที่ Predict
# -------------------------
filtered_labels = [int(label.cpu().item()) for label, score in zip(labels, scores) if score > threshold]
class_counts = Counter(filtered_labels)

print("จำนวนแต่ละ class:", class_counts)
print("จำนวนทั้งหมด:", sum(class_counts.values()))

# -------------------------
# พล็อตจำนวน Predict (Bar Chart)
# -------------------------
plt.figure(figsize=(6,4))
plt.bar(class_counts.keys(), class_counts.values(), color='skyblue')
plt.xlabel("Class ID")
plt.ylabel("Count")
plt.title("Number of Predictions per Class")
plt.show()


```
